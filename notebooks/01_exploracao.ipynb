```python
import pandas as pd

pd.set_option("display.max_columns", None)
pd.set_option("display.max_colwidth", 50)
```

```python
#carregar o cadastro de fundos

cad = pd.read_csv(
    "../data/raw/cad_fi.csv",
    sep=";",
    encoding="latin1",
    low_memory=False
)

print(f"Linhas: {len(cad)}")
print(f"Colunas: {cad.shape[1]}")
print(cad.columns.tolist())
```

```python
#ver todas as colunas com descrição

print(cad.dtypes.to_string())
```

```python
#amostra

cad.head(3)
```

```python
#carregar um mês de informe diário

inf = pd.read_csv(
    "../data/raw/inf_diario_fi_202501.csv",
    sep=";",
    encoding="latin1",
    low_memory=False
)

print(f"Linhas: {len(inf)}")
print(f"Colunas: {inf.shape[1]}")
print(inf.columns.tolist())
inf.head(3)
```

```python
#verificar nulos nas colunas de relacionamento

cols = ['CNPJ_ADMIN', 'CPF_CNPJ_GESTOR', 'CLASSE', 
        'CNPJ_AUDITOR', 'CNPJ_CUSTODIANTE', 'CNPJ_CONTROLADOR']

print("Valores nulos por coluna:")
print(cad[cols].isnull().sum())
print()
print("Percentual nulo:")
print((cad[cols].isnull().sum() / len(cad) * 100).round(1))
```

```python
#verificar situação dos fundos com nulos

print(cad['SIT'].value_counts())
print()
print("Nulos em CNPJ_ADMIN por situação:")
print(cad.groupby('SIT')['CNPJ_ADMIN'].apply(lambda x: x.isnull().sum()))
```

```python
#cruzar fundos ativos com quem tem informe diário em 2025

# quantos CNPJs únicos aparecem nos informes de janeiro
fundos_inf = set(inf['CNPJ_FUNDO_CLASSE'].unique())

# quantos CNPJs do cadastro estão nos informes?
fundos_cad = set(cad['CNPJ_FUNDO'].unique())

apenas_inf = fundos_inf - fundos_cad
apenas_cad = fundos_cad - fundos_inf
em_ambos   = fundos_inf & fundos_cad

print(f"Fundos com informe em jan/2025:        {len(fundos_inf)}")
print(f"Fundos no cadastro:                    {len(fundos_cad)}")
print(f"Fundos em ambos (cruzamento):          {len(em_ambos)}")
print(f"Fundos no informe mas não no cadastro: {len(apenas_inf)}")
print(f"Fundos no cadastro mas não no informe: {len(apenas_cad)}")
```

```python
#comparar formato dos CNPJs

print("Exemplo CNPJ no cadastro:")
print(cad['CNPJ_FUNDO'].head(3).tolist())

print("\nExemplo CNPJ no informe:")
print(inf['CNPJ_FUNDO_CLASSE'].head(3).tolist())
```

```python
#verificar se um CNPJ do informe existe no cadastro

# Pega um CNPJ do informe que NÃO está no cadastro
cnpj_teste = list(apenas_inf)[0]
print(f"CNPJ do informe não encontrado no cadastro: {cnpj_teste}")

# Tenta buscar no cadastro
resultado = cad[cad['CNPJ_FUNDO'] == cnpj_teste]
print(f"Encontrado no cadastro: {len(resultado)} linha(s)")

# Verifica a situação dos fundos que ESTÃO em ambos
situacao_comuns = cad[cad['CNPJ_FUNDO'].isin(em_ambos)]['SIT'].value_counts()
print(f"\nSituação dos {len(em_ambos)} fundos em comum:")
print(situacao_comuns)
```

```python
#explorar o campo ID_SUBCLASSE

print("Valores únicos de TP_FUNDO_CLASSE:")
print(inf['TP_FUNDO_CLASSE'].value_counts())
print()
print("Nulos em ID_SUBCLASSE:")
print(inf['ID_SUBCLASSE'].isnull().sum())
print()
print("Exemplo de ID_SUBCLASSE:")
print(inf['ID_SUBCLASSE'].dropna().head(5).tolist())
```

```python
#testar cruzamento só com fundos FI

inf_fi = inf[inf['TP_FUNDO_CLASSE'] == 'FI']

fundos_fi = set(inf_fi['CNPJ_FUNDO_CLASSE'].unique())
fundos_cad_ativos = set(cad[cad['SIT'] != 'CANCELADA']['CNPJ_FUNDO'].unique())

em_ambos_fi = fundos_fi & fundos_cad_ativos

print(f"CNPJs únicos no informe (tipo FI):     {len(fundos_fi)}")
print(f"CNPJs no cadastro (não cancelados):    {len(fundos_cad_ativos)}")
print(f"Cruzamento:                            {len(em_ambos_fi)}")
```

```python
#explorar o cadastro de classes

classe = pd.read_csv(
    "../data/raw/registro_classe.csv",
    sep=";",
    encoding="latin1",
    low_memory=False
)

print(f"Linhas: {len(classe)}")
print(f"Colunas: {classe.shape[1]}")
print(classe.columns.tolist())
classe.head(3)
```

```python
#explorar registro de fundos e testar cruzamento

fundo = pd.read_csv(
    "../data/raw/registro_fundo.csv",
    sep=";",
    encoding="latin1",
    low_memory=False
)

print(f"=== registro_fundo.csv ===")
print(f"Linhas: {len(fundo)}")
print(f"Colunas: {fundo.columns.tolist()}")
print()

# Testar formato do CNPJ_Classe vs informe
print("=== Formato CNPJ ===")
print("CNPJ_Classe (registro):", classe['CNPJ_Classe'].head(3).tolist())
print("CNPJ no informe:", inf['CNPJ_FUNDO_CLASSE'].head(3).tolist())
```

```python
#padronizar CNPJs e testar cruzamento

# Formatar CNPJ do registro para o padrão do informe (XX.XXX.XXX/XXXX-XX)
def formatar_cnpj(cnpj):
    cnpj = str(int(cnpj)).zfill(14)
    return f"{cnpj[:2]}.{cnpj[2:5]}.{cnpj[5:8]}/{cnpj[8:12]}-{cnpj[12:]}"

classe['CNPJ_Classe_fmt'] = classe['CNPJ_Classe'].apply(formatar_cnpj)

# Testar cruzamento
cnpjs_classe = set(classe['CNPJ_Classe_fmt'])
cnpjs_inf    = set(inf['CNPJ_FUNDO_CLASSE'])

em_ambos = cnpjs_classe & cnpjs_inf

print(f"CNPJs únicos no registro_classe:  {len(cnpjs_classe)}")
print(f"CNPJs únicos no informe (jan/25): {len(cnpjs_inf)}")
print(f"Cruzamento:                       {len(em_ambos)}")
print()
print("Exemplo CNPJ formatado:", classe['CNPJ_Classe_fmt'].head(3).tolist())
```
